```python
t = data['verification_time']

data['year_column'] = pd.DatetimeIndex(t).year

data.head(100)
```

```python
#2 - Build a time series plot of phishing attacks by year
#All data points are considered for this one
ax = data['year_column'].value_counts().plot(kind = 'line', figsize = (14,8), title = "Phishing Attacks by Each Year", color='#334f8d',fontsize=11, zorder=4)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
# 3 - Find the brand that has been attacked the most in this period and plot a time series chart.
data_top = data.query('target == "Other"')
data_top.set_index('year_column')
```

```python

ax = data_top['year_column'].value_counts().plot(x='target',kind = 'line', figsize = (14,8), title = "Phishing Attacks by Each Year", color='#334f8d',fontsize=11, zorder=4)
ax.set(title='Time series chart of phishing attacks for the most attacked brand',
       ylabel='# of Phishing Attacks')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
## 5 - Any interesting insights you can draw from the phishing_url column?
url_part = data['url'].str.split("/")
#print (url_part)
host=[]
for part in url_part:
    host.append(part[2])
    
data['host'] = host
```

```python
evil_attackers = data.groupby('host').count()[['phish_id']].sort_values(by = ['phish_id'], ascending = False).head(10)
print(evil_attackers)

## Looks like most frequent phishing attacks are the ones people are most likely to click at - The host of phishing attacks from google and microsoft.
```
